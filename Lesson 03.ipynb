Времена года (4)
Написать функцию season, принимающую 1 аргумент — номер месяца (от 1 до 12), и возвращающую время года, которому этот месяц принадлежит (зима, весна, лето или осень).

```python
def season_or(n):
  if n==1 or n==2 or n==12:
    return "зима"
  elif n==3 or n==4 or n==5:
    return "весна"
  elif n==6 or n==7 or n==8:
    return "лето"
  elif n==9 or n==10 or n==11:
    return "осень"
  else:
    return "неправильное значение"

season_or(12)
```

```python
def season_tuple(n):
    winter = (12, 1, 2)
    spring = (3, 4, 5)
    summer = (6, 7, 8)
    fall = (9, 10, 11)

    if n in winter:
        return "зима"
    elif n in spring:
        return "весна"
    elif n in summer:
        return "лето"
    elif n in fall:
        return "осень"
    else:
        return "неправильное значение"

season_tuple(12)
```

Банковский вклад (5)
Пользователь делает вклад в размере a рублей сроком на years лет под 10% годовых (каждый год размер его вклада увеличивается на 10%. Эти деньги прибавляются к сумме вклада, и на них в следующем году тоже будут проценты).

Написать функцию bank, принимающая аргументы a и years, и возвращающую сумму, которая будет на счету пользователя.

```python
def bank(a, years):
  for i in range(years):
    a = int(a+10*a/100)
  return a

bank(1000, 5)
```

```python
def bank2(a, years):
  return a * 1.1 ** years

bank2(1000, 5)
```

```python
%%timeit
bank(1000, 5)
```

```python
%%timeit
bank2(1000, 5)
```

Простые числа (6)
Написать функцию is_prime, принимающую 1 аргумент — число от 0 до 1000, и возвращающую True, если оно простое, и False - иначе.

```python
def is_prime(a):
  for i in range(a):
    if(a % 2 == 0):
      return "Число сложное"
    else:
      return "Число простое"

is_prime(6), is_prime(9)
```

```python
def is_prime(a):
    for i in range(2, a):
        if a % i == 0:
            return False

    return True

is_prime(6), is_prime(9), is_prime(1), is_prime(2), is_prime(17), is_prime(3)
```
